```python
import pandas as pd
import datetime
```

```python
citibike_csv = '201906_tripdata.csv'
# Read with Pandas
citibike_df = pd.read_csv(citibike_csv)
citibike_df.head()
citibike_df
```

```python
list(citibike_df)
```

```python
age = pd.Timestamp('now').year - citibike_df['birth year'] 
age
```

```python
citibike_df['age'] = age
```

```python
list(citibike_df)
```

```python
citibike_df.head()
```

```python
del citibike_df['birth year']
```

```python
citibike_df.head()
```

```python
citibike_df = citibike_df.drop(citibike_df[citibike_df.gender == 0].index)
```

```python
citibike_df
```

```python
citibike_df['gender'] = citibike_df['gender'].astype('str')
```

```python
citibike_df['gender'].replace({'1': 'M', '2': 'F'}, inplace=True)
citibike_df.head()
```

```python
citibike_df.to_csv('citi_201906.csv', index=False)
```
